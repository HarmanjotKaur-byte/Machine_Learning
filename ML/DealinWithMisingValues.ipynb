```python
import numpy as np
import pandas as pd
```

```python
d={"Apple":[45,34,np.nan,15,np.nan,65],
  "Orange":[34,12,56,95,np.nan,30],
  "Grapes":[1,2,5,6,9,8],
 "Papaya":[10,np.nan,5,6,9,8]
}#np.nan refers to khali space not even 0
d
```

```python
df=pd.DataFrame(d)
df
```

```python
#isnull() function:- return the result in True/False
df.isnull()
```

```python
#isnull().sum() function:- return the no of null values in respective columns
df.isnull().sum()
```

```python
#notnull() function
df.notnull()
```

```python
#notnull().sum()
df.notnull().sum()
```

```python
#1.fillna -> better for small dataset but not for large
df.fillna(0)
```

```python
df.fillna(100)
```

```python
#deprecated way but do same as bfill()
df.fillna(method="pad")
```

```python
df.ffill()  #pichliyan values agge on giyan
```

```python
df.bfill()  #agliyan values piche on giyan
```

```python
d={"Apple":[45,34,np.nan,15,np.nan,65],
  "Orange":[34,12,56,95,np.nan,30],
  "Grapes":[1,2,5,6,9,8],
 "Papaya":[10,np.nan,5,6,9,8]
}
d
```

```python
df=pd.DataFrame(d)
```

```python
#axis=0 for row-wise 
#we will get row which donot have nan val
print(df)
df.dropna(axis=0)
```

```python
#axis=1 for columns
#we will get column which donot have nan val
print(df)
df.dropna(axis=1)
```
