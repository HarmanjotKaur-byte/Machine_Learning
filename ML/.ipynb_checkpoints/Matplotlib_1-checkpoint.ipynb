```python
#importing
#way 1
import matplotlib.pyplot as plt
#way 2
from matplotlib import pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
x=np.array([1,2,3,5])
plt.plot(x)
plt.show()
#plt.plot(x) will show address as well as line graph
#plt.show() will only show the graph and no address
#plt.plot(x) here since we have passed only one thing that is array so array values will be on y- axis whereas indices (0,1,2,3) will be on x- axis
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y)  #here first argument is x axis and second argument is y axis
```

```python
#for highlighting intersection points from data
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="o") 
```

```python
#for desired color line
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="o",color="r") 
```

```python
#for changing marker edge color (mec)
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="o",color="r",mec="g") 
```

```python
#for changing shape of marker (using H) [here hexagon has flat shape above]
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="H",color="r",mec="g") 
```

```python
#for changing shape of marker (using h) [here hexagon has point up shape]
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="r",mec="g") 
```

```python
#for size of marker
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="r",mec="g",markersize=10) 
```

```python
#for size of marker
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="H",color="r",mec="g",markersize=10) 
```

```python
#for changing line style
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="r",mec="g",linestyle="--") #or linestyle="dashed" 
```

```python
#for line width
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="r",mec="g",linewidth=6) 
```

```python
#for marker face color(mfc)
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="y",mec="r",mfc="b") 
```

```python
#for giving labels(names) to x axis , y axis and table name
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="y",mec="r",mfc="b") 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Graph")
```

```python
#for grids
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="y",mec="r",mfc="b") 
plt.grid()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="y",mec="r",mfc="b") 
plt.grid(axis="x")
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.plot(x,y,marker="h",color="y",mec="r",mfc="b") 
plt.grid(axis="y")
```
