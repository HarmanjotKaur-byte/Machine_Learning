```python
#for colored map pof correlation
sns.heatmap(df.corr(numeric_only=True))
```

```python
import seaborn as sns
sns.dogplot() #reuires internet connection
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#linear reg->Best fit line (to predict future o/p)
#scatter shadow due to shadow [distance b/w points and line]

#kde:-kernel density estimation 
```

```python
#hue='sex' for separating on gender basis
sns.pairplot(df,hue='sex')
```

```python
#with kde curve
sns.distplot(df['tip'])
```

```python
#without kde curve
sns.distplot(df['tip'],kde=False)
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#1.countplot :- to see no of categorical column
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#2->Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
# ------ represents outlier
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
#Box plot:- (whisker plot) represents information from five number summary
sns.boxplot(x='smoker',y='total_bill',data=df)
#circles above --------- line repsresents outlier
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(y='day',x='total_bill',data=df,hue='smoker')
```

```python
sns.violinplot(y='day',x='total_bill',data=df,palette='rainbow')
```

```python
sns.violinplot(y='day',x='total_bill',data=df,palette='rainbow',hue='smoker')
```
