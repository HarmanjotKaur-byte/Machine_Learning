---------------------------------------------------- Scatter Plot ---------------------------------------------------------------------
#Scatter plot: Find corelation b.w two variables

```python
import pandas as pd  #good to import this lib as it can be used
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,5,6,7,6,12,19,10,40])
y=np.array([10,20,33,54,64,75,6,12,19,90,40])
plt.scatter(x,y)         #bydeafult color is blue
```

```python
#if we want color of dots to be red [specific]
x=np.array([1,2,3,5,6,7,6,12,19,10,40])
y=np.array([10,20,33,54,64,75,6,12,19,90,40])
plt.scatter(x,y,color="red")         
```

```python
x=np.array([1,2,3,5,6,7,6,12,19,10,40])
y=np.array([10,20,33,54,64,75,6,12,19,90,40])
plt.scatter(x,y,color="green")         
x=np.array([11,234,35,56,67,77,68,120,199,107,406])
y=np.array([103,204,335,546,646,757,67,127,195,904,403])
plt.scatter(x,y,color="red")         
```

```python
#for different colors for each intersection
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=["blue","pink","green","yellow","violet","red","lavender","black"]
plt.scatter(x,y,color=colors)    # or write c=colors
```

```python
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=np.array([120,20,23,58,56,89,47,12])
plt.scatter(x,y,c=colors)   
```

```python
#use of color maps
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=np.array([120,20,23,58,56,89,47,12])
plt.scatter(x,y,c=colors,cmap="viridis")   
```

```python
from matplotlib import colormaps
list(colormaps)
```

```python
#use of colorbar
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=np.array([120,20,23,58,56,89,47,12])
plt.scatter(x,y,c=colors,cmap="viridis")   
plt.colorbar()
```

```python
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=np.array([120,20,23,58,56,89,47,12])
plt.scatter(x,y,c=colors,cmap="plasma")   
plt.colorbar()
```

```python
#increasing size of markers 
x=np.array([1,2,3,5,6,7,6,12])
y=np.array([10,20,33,54,64,75,6,45])
colors=np.array([120,20,23,58,56,89,47,12])
sizes=np.array([10,34,56,67,89,450,67,89])
plt.scatter(x,y,c=colors,cmap="plasma",s=sizes)   
plt.colorbar()
```

```python
x=np.random.randint(100,size=100) 
y=np.random.randint(100,size=100) 
colors=np.random.randint(100,size=100)
sizes=5*np.random.randint(100,size=100)
plt.scatter(x,y,c=colors,s=sizes,cmap="plasma")
plt.colorbar()
```

```python
x=np.random.randint(100,size=100) 
y=np.random.randint(100,size=100) 
colors=np.random.randint(100,size=100)
sizes=5*np.random.randint(100,size=100)
plt.scatter(x,y,c=colors,s=sizes,cmap="plasma",alpha=0.3) 
#alpha value ranges from 0 to 1 
# [0->More transparent
#1->Less transparent more opaque]
plt.colorbar()
```

---------------------------------------------    Bar Graph     ----------------------------------------------------------

```python
x=np.array(["A","B","C","D"])
y=np.array([12,45,52,13])
plt.bar(x,y,color="pink") #same color to all
```

```python
#Diff colors for diff bars
#Vertical bar
x=np.array(["A","B","C","D"])
y=np.array([12,45,52,13])
colors=["red","yellow","blue","pink"]
plt.bar(x,y,color=colors)
```

```python
#Diff colors for diff bars
#Vertical bar
x=np.array(["A","B","C","D"])
y=np.array([12,45,52,13])
colors=["red","yellow","blue","pink"]
plt.bar(x,y,color=colors,width=0.6)   #can control width not height
```

```python
#Horizontal bar : can control height not width
x=np.array(["A","B","C","D"])
y=np.array([12,45,52,13])
colors=["red","yellow","blue","pink"]
plt.barh(x,y,color=colors,height=0.4) #just write h
```
