```python
#Importing onehotencoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#converting dictionary to dataframe
d={"fruits":["apple","banana","orange","apple","banana","orange"],
   "color":["Red","Yellow","Orange","Red","Yellow","Orange"],
   "size":[10,20,30,40,50,89]
  }
```

```python
d
```

```python
df=pd.DataFrame(d)
df
```

```python
#Since fruits amd color are textual we cant feed them directly to ML model
#we need to convert textual to numeric
```

```python
#Creating the object of onehotencoder (now we can use encoder.fxn_name)
encoder=OneHotEncoder()
```

```python
#fit->Not using value
#fit-tranform->Using values

encoder.fit_transform(df[["fruits"]])
```

```python
encoder.fit_transform(df[["fruits"]]).toarray()
```

```python
encoded=encoder.fit_transform(df[["fruits"]])
```

```python
encoded
```

```python
x=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
x
```

```python
print(df)
pd.concat((df,x),axis=1)
```

```python
encoded_2=encoder.fit_transform(df[["color"]])
encoded_2
```

```python
y=pd.DataFrame(encoded_2.toarray(),columns=encoder.get_feature_names_out())
y
```

```python
print(df)
pd.concat((df,x,y),axis=1)
```

```python
#Importing ordinalencoder 
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

```python
#converting dictionary to dataframe
d={"Education":["Phd","PG","UG","School"],
   "Review":["Excellent","VeryGood","Good","Poor"],
   "Result":["Yes","No","Yes","No"]
  }
d
```

```python
df=pd.DataFrame(d)
df
#ed, rev -> ordinal 
#result -> label
```

```python
ord=OrdinalEncoder(categories=[["School","UG","PG","Phd"],
                               ["Poor","Good","VeryGood","Excellent"]]
                               )
```

```python
x=["Education","Review"]
```

```python
df[x]=ord.fit_transform(df[x])
```

```python
print(df)
```

```python
le=LabelEncoder()
df["Result"]=le.fit_transform(df["Result"])
df
```
